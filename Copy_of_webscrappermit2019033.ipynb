```
main_data = []
#print(data)
pages = [str(i) for i in range(1,7)]
for page in pages :
  url = "https://www.oyorooms.com/hotels-in-manali/?page="+page
  response = requests.request("GET", url)
  #sleep(randint(1,5))
  #print(response)
  data = BeautifulSoup(response.text, 'html.parser')
  data2 = data.find_all('div',attrs = {'class' :'oyo-cell--12-col oyo-cell--8-col-tablet oyo-cell--4-col-phone'})
  #print(data2)
  for j in data2:
    title = j.find('h3')
    #print(title.text)
    rating = j.find('div',attrs = {'class': 'hotelRating'}).find('span').text
    price = j.find('div',attrs = {'class' : 'listingPrice__numbers'}).text.strip('₹').split("₹")[0]
    price = str(j.find('span',attrs = {'class': 'listingPrice__finalPrice'}).text.strip('₹'))
    #print(price)
    dta = {}
    dta['name'] = title.text
    dta['rating'] = rating
    dta['price'] = price
    main_data.append(dta)
    print(dta)
```

```
dataframe = pd.DataFrame.from_dict(main_data)
print(dataframe)
```
